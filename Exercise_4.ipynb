**Q4.** From the data generated in task no. 1, extract Day, Month and Year from the column Order.Date and create
separate columns namely Day, Month and Year and store it in the same dataset only

```r
library(lubridate)
i <- dmy(e$Order.Date)
j <- list(day(i), month(i), year(i))
k <- cbind(e, j)
names(k)[11] <- "Day"
names(k)[12] <- "Month"
names(k)[13] <- "Year"
head(k)
```

**Q5.** From the data generated in task no. 1, create another column which stores the original date one year later
named After1Year. E.g. if date is 23 rd June 2014 then the date in the created column should be 23 rd June
2015.

```r
l <- dmy(e$Order.Date) + years(1)
m <- format(l, "%d-%b-%y")
n <- cbind(e, m)
names(n)[11] <- "After1Year"
n
```
